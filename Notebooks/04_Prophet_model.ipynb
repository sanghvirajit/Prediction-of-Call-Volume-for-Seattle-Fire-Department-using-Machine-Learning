```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib import pyplot
import calendar 
import datetime
import seaborn as sns
%matplotlib inline
```

# Hourly Time Series Forecasting using Facebook's Prophet

```python
data = pd.read_csv("../Data/hourly_data.csv")
data.head()
```

```python
data['datetime'] = data[['year', 'month', 'day_of_month', 'hour']].apply(lambda s : datetime.datetime(*s),axis = 1)

del data['year']
del data['month']
del data['day_of_month']
del data['hour']

# rearranging the columns
cols = ['datetime', 'call_counter']
data = data[cols]

data.head()
```

```python
# hourly call counter for first month
plt.plot(data.index[:744], data['call_counter'][:744])
plt.show()
```

```python
data = data.rename(columns={"datetime": "ds", "call_counter": "y"})
data.head()
```

```python
split_date = '01-01-2022'

data_train = data.loc[data['ds'] <= split_date].copy()
data_test = data.loc[data['ds'] > split_date].copy()
```

```python
data_train.head()
```

# Prophet Model

```python
from fbprophet import Prophet
```

```python
model = Prophet()
```

```python
model.fit(data_train)
```

```python
y_predict = model.predict(df=data_test)
```

```python
y_predict.head()
```

```python
# Plot the forecast
model.plot(y_predict)
plt.xlim([datetime.date(2021, 1, 1), datetime.date(2022, 2, 1)])
plt.show()
```

```python
model.plot_components(y_predict);
```

```python
f, ax = plt.subplots(1)
f.set_figheight(5)
f.set_figwidth(15)
ax.scatter(data_test['ds'], data_test['y'], color='r', label = 'True')
fig = model.plot(y_predict, ax=ax)
plt.legend()
plt.xlim([datetime.date(2021, 12, 1), datetime.date(2022, 2, 1)])
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

RMSE_test = sqrt(mean_squared_error(y_true=data_test['y'], y_pred=y_predict['yhat']))

print("RMSE on testing data: {}".format(RMSE_test))
```

# Actual (measured) call volume vs. predicted call volume for a month

```python
# Plot the forecast with the actuals
f, ax = plt.subplots(1)
f.set_figheight(5)
f.set_figwidth(15)
ax.scatter(data_test['ds'], data_test['y'], color='r', label = 'True')
fig = model.plot(y_predict, ax=ax)
plt.legend()
plt.xlim([datetime.date(2022, 1, 1), datetime.date(2022, 2, 1)])
plt.suptitle('January 2022 Forecast vs Actuals')
```
