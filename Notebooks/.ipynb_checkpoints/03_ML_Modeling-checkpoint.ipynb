# Data Splitting

we will train the model for last 4 year (2017 to 2020), validate on 2021 and test/forecast for year 2022

The training set is used to fit the models and tune the hyperparameters; the validation set is used to estimate prediction error for model selection; the test set is used for assessment of the generalization error of the final chosen model.

```python
X_train = data[data['year'] < 2021]
y_train = X_train['call_counter']

del X_train['call_counter']
X_train = X_train.set_index('date')

X_val = data[data['year'] == 2021]
y_val = X_val['call_counter']

del X_val['call_counter']
X_val = X_val.set_index('date')

X_test = data[data['year'] == 2022]
y_test = X_test['call_counter']

del X_test['call_counter']
X_test = X_test.set_index('date')
```

### Hyperparameter tuning with randomsearchcv()

```python
def bestparam_randCV(model,hyperparam,x_train, y_train, n_iter=1000):
    
    hyperparam = hyperparam
    randomizedCV = RandomizedSearchCV(model, param_distributions = hyperparam, cv = 10,
                                      n_iter = n_iter, scoring = 'neg_mean_squared_error', n_jobs=-1, 
                                      random_state = 42, verbose = True)
    
    randomizedCV.fit(x_train, y_train)
    
    #print (randomizedCV.cv_results_)
    print ('Best MSE', randomizedCV.score(x_train, y_train))
    print ('Best Param', randomizedCV.best_params_)
    return randomizedCV
```

# Forecasting for year 2022 with LightGBM

Traning the final model for last 5 years (2017-2021) and forecasting for the year 2022

```python
X_train = data[data['year'] < 2022]
y_train = X_train['call_counter']

del X_train['call_counter']
X_train = X_train.set_index('date')

X_test = data[data['year'] == 2022]
y_test = X_test['call_counter']

del X_test['call_counter']
X_test = X_test.set_index('date')
```

```python
final_model = lgb.LGBMRegressor(
                       num_leaves = 30,
                       max_depth = 6,
                       learning_rate = 0.1
)

lightgbm_final_fit = final_model.fit(X_train, y_train)
```

```python
lightgbm_final_predict = lightgbm_final_fit.predict(X_test)
lightgbm_final_predict = lightgbm_final_predict.round()
```

```python
df = X_test.copy()
df['true'] = y_test.to_numpy()
df['predict'] = lightgbm_final_predict
df.insert(0,'datetime','')
df['datetime'] = df[['year', 'month', 'day_of_month', 'hour']].apply(lambda s : datetime.datetime(*s),axis = 1)
df.head()
```

# Actual (measured) call volume vs. predicted call volume for a month

```python
# Plot the forecast with the actuals
f, ax = plt.subplots(1)
f.set_figheight(5)
f.set_figwidth(15)
ax.scatter(df['datetime'], df['true'], color='r', label = 'Actual')
ax.scatter(df['datetime'], df['predict'], color='b', label = 'Predicted')
plt.legend()
plt.xlim([datetime.date(2022, 1, 1), datetime.date(2022, 2, 1)])
plt.suptitle('Actual (measured) call volume vs. predicted call volume for a month of January 2022')
```

# Random Date and time

```python
model = lgb.LGBMRegressor(
                       num_leaves = 30,
                       max_depth = 6,
                       learning_rate = 0.1
    )
```

```python
date_time_dict = {
    'datetime': '2022-01-01 04:00:00'
}
```

```python
def train(model, X_train, y_train):
    
    model = model.fit(X_train, y_train)
    
    return model
```

```python
def prepare_input(date_time_dict):
    
    df_date_time =  pd.DataFrame([date_time_dict])
    df_date_time['datetime'] = pd.to_datetime(df_date_time['datetime'])
    
    return df_date_time
```

```python
def create_features(df, label=None):
    
    df["year"]           = df["datetime"].apply(lambda time: time.year)
    df["day_of_month"]   = df["datetime"].apply(lambda time: time.day)
    df["month"]          = df["datetime"].apply(lambda time: time.month)
    df["hour"]           = df["datetime"].apply(lambda time: time.hour)
    df['day_of_year']    = df['datetime'].apply(lambda time: time.dayofyear)
    df['day_of_week']    = df['datetime'].apply(lambda time: time.dayofweek)
    df['week_of_year']   = df['datetime'].apply(lambda time: time.weekofyear)
    
    del df['datetime']
    
    return df
```

```python
def predict(model, data):

    predict = model.predict(data).round()
    
    return predict[0]
```

```python
model = train(model, X_train, y_train)
```

```python
df_date_time = prepare_input(date_time_dict)
```

```python
df_date_time = create_features(df_date_time)
```

```python
y_predict = predict(model, df_date_time)
y_predict
```
